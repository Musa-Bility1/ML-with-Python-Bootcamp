Step 1: Loadingh & Perserving

```python
import pandas as pd

# 1. The data link
url = "https://raw.githubusercontent.com/Musa-Bility1/ML-with-Python-Bootcamp/refs/heads/main/market_dirty.csv"

# 2. Read the data
df = pd.read_csv(url)

# 3. Keep a raw copy for comperison
df_raw = df.copy()


# 4. Look at the first few row
print(df.head())
```

Step 2: Asking the Data Questions

```python
df.info()
df.isna().sum()
```

Step 3: Removing Blank Prices

```python
# Drop row where price si empty
df = df.dropna(subset=['price_lrd'])

# See how many records survived
print(f'Original: {len(df_raw)}')
print(f'Clean: {len(df)}')
```

Step 4: Filling in the Gaps

```python
df = df_raw.copy()
df ['price_lrd'] = df['price_lrd'].fillna(df['price_lrd'].median())
print(df)
```

Step 5: The "Speed Limit-Removing Outlayer

```python
# Find the speed limit(Top 5% doundary)
limit = df['price_lrd'].quantile(0.95)

# Bring the crazy spikes down to the limit
df['price_lrd'] = df['price_lrd'].clip(upper=limit)
print(df['price_lrd'].round(2))
```
