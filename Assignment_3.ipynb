QUESTION 1: Rice Market Analysis (NumPy)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# a) Create a NumPy array
prices = np.array([1200, 1250, 1180, 1300, 1220, 1270, 1240])
print("Q1a) NumPy array:", prices)

# b) Mean price
mean_price = np.mean(prices)
print(f"Q1b) Mean price: {mean_price:.2f} LRD")

# c) Highest and lowest prices
highest = np.max(prices)
lowest = np.min(prices)
print(f"Q1c) Highest price: {highest} LRD | Lowest price: {lowest} LRD")

# d) Difference between highest and lowest
price_range = highest - lowest
print(f"Q1d) Difference (range): {price_range} LRD")

# e) Explanation
print("""
Q1e) What the mean tells the seller:
     The mean price of {:.2f} LRD represents the typical price per bag
     across the seven days. The seller can use this average as a
     benchmark to judge whether a given day's price is above or below
     normal, helping with buying/selling decisions.
""".format(mean_price))
```

QUESTION 2: Price Shock Investigation (NumPy)

```python
# a) Update last price and recalculate mean
prices_new = prices.copy()
prices_new[-1] = 2000
new_mean = np.mean(prices_new)
print(f"Q2a) New mean price: {new_mean:.2f} LRD")

# b) Compare old and new mean
print(f"Q2b) Old mean: {mean_price:.2f} LRD | New mean: {new_mean:.2f} LRD")
print(f" Difference: +{new_mean - mean_price:.2f} LRD")

# c) Explanation
print("""
Q2c) Effect of an unusually high value:
     The single spike from 1240 to 2000 LRD raised the mean by
     {:.2f} LRD. This shows that the arithmetic mean is sensitive
     to outliers (extreme values). A data analyst must check for
     such values because they can distort the average and lead to
     poor business or policy decisions.
""".format(new_mean - mean_price))
```

QUESTION 3: County Crop Survey (Pandas)

```python
# a) Loadung the dataset
df = pd.read_csv("https://raw.githubusercontent.com/Musa-Bility1/ML-with-Python-Bootcamp/refs/heads/main/cassava_data.csv")
print("Q3a) Dataset loaded successfully.\n")

# b) First five rows
print("Q3b) First five rows:")
print(df.head(), "\n")

# c) Shape
print(f"Q3c) Rows: {df.shape[0]} | Columns: {df.shape[1]}\n")

# d) Column names
print("Q3d) Column names:", df.columns.tolist(), "\n")

# e) Statistical summary
print("Q3e) Statistical summary:")
print(df.describe(), "\n")
```

QUESTION 4: Finding Areas with Low Production (Pandas)

```python
# a) Counties with cassava_bags < 80
low_production = df[df["cassava_bags"] < 80]
print("Q4a) Counties with cassava_bags < 80:")
print(low_production[["county", "cassava_bags"]], "\n")

# b) Count
count_low = len(low_production)
print(f"Q4b) Number of counties: {count_low}\n")

# c) Explanation
print("""
Q4c) How this helps an agricultural officer:
     Identifying counties with low cassava production (< 80 bags)
     allows an agricultural officer to:
       - Target those areas for extension services and training.
       - Allocate fertiliser subsidies or improved seed varieties.
       - Investigate root causes (poor rainfall, few farmers, pests).
       - Track whether interventions improve output in future seasons.
""")
```

QUESTION 5: Visualizing School Attendance (Matplotlib)

```python
attendance = [35, 37, 40, 38, 42, 39, 41]
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# a, b, c) Line graph with title, labels, and markers
plt.figure(figsize=(9, 5))
plt.plot(days, attendance, marker='o', color='steelblue',
         linewidth=2, markersize=8, markerfacecolor='orange')

# b) Title and axis labels
plt.title("Daily School Attendance Over One Week", fontsize=14, fontweight='bold')
plt.xlabel("Day of the Week")
plt.ylabel("Number of Students")
plt.xticks(rotation=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("q5_attendance.png", dpi=120)
plt.show()

print("""
Q5d) Trend description:
     The graph shows a generally increasing trend in attendance from
     Monday (35) to Friday (42), with a slight dip on Thursday (38).
     Saturday dips to 39 before recovering slightly to 41 on Sunday.
     Overall, attendance is rising across the week, which may indicate
     growing student engagement or an upcoming exam/event.
""")
```
